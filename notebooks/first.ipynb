```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

data = yf.download("ZW=F", start = "2015-01-01", end = "2026-01-30")
data['Close'].plot(title='Wheat Futures Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid()
plt.show()

```

```python
import os

folder_path = 'Project_commodities/data'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

data.to_csv(f'{folder_path}/wheat_universe.csv')
```

```python
proxies = ["DBA", "JJG", "VEGI"]
stocks = ["ADM", "BG", "CTVA", "NTR", "DE", "MOS", "CF", "FMC", "AGCO", "WBC"]

all_tickers = proxies + stocks

data2 = yf.download(all_tickers, start="2015-01-01", end="2026-01-30")
data2['Close'].plot(title='Proxies and Stocks Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid()
plt.legend(all_tickers)
plt.show()
```
